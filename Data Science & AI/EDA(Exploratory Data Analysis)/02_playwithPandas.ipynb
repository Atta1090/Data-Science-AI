# **how to rename columns**

```python
df  = pd.DataFrame({'A col': [1, 2, 3, 4,5,6,7,7], 'B col': [4, 5, 6,4,5,6,7,7]})
df
```

```python
df.rename(columns={'A col': 'A', 'B col': 'B'}, inplace=True)
```

```python
df
```

```python
df.columns = ['AA', 'BB']
```

```python
# to replace any character in a string
df.columns = df.columns.str.replace('AA', 'A')
```

```python
df
```

```python
# adding prefix in column
df = df.add_prefix('col_')
df
```

```python
# adding a suffix to the column name
df.add_suffix('_new')
```

# **8 Types Casting: Convert string to int**

```python
df = pd.DataFrame({'col_ a':["1","2",'3','4'],
                   'col_b':[5,6,7,8]})
df
```

```python
df.dtypes
```

```python
df.columns
```

```python
df.astype({'col_ a': 'int'})
```

```python
df.astype({'col_ a': 'float'})
```

```python
# another way to do it
# errorrs='coerce' will convert non-numeric values to NaN
# When converting data to numeric using pd.to_numeric(),
#  not all values may be valid numbers (e.g., strings like "abc", "N/A", "--", etc.). 
# By default, pandas would raise an error if it encounters such invalid entries.
pd.to_numeric(df['col_ a'], errors='coerce')
```
