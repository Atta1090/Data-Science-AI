# **Basic Data Structures in Pandas**

Panadas Provides two types of classes for handling data.
1. **Series:** A one-dimensional labeled array capable of holding any data type (integer, string, float, python objects etc)
2. **DataFrame:** A two-dimensional table of values with column labels with row labels.

## **1. Object Creation**

```python
# Object Creation
a = pd.Series([1,2,3,6,8])
a
```

```python
# creatinga a Data frame using pandas
dates = pd.date_range("20130101", periods = 6)
dates
```

```python
# creating a data frame a by passing a dictionary of objects
import numpy as np
df2 = pd.DataFrame(
    {
        "A": 1.0,
        "B": pd.Timestamp("20130102"),
        "C": pd.Series(1, index=list(range(4)), dtype="float32"),
        "D": np.array([3] * 4, dtype="int32"),
        "E": pd.Categorical(["test", "train", "test", "train"]),
        "F": "foo",
    }
)
df2
```

## **7. Missing Data**

```python
# how to drop rows
# df.dropna()  # drop rows with missing values
df.dropna(how= 'any')
# fills missing values
df.fillna(0)  # fills missing values with 0
df.fillna(value=5)
# isna() gets the boolean mask where values are nan:
pd.isna(df)
```

## **8. Operations**


```python
# stats
# caluclate mean value for each column
df.mean()
# mean value for each row
df.mean(axis=1)
# median
df.median()
# mode
df.mode()
# standard deviation
df.std()
# variance
df.var()
```
